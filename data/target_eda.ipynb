```python

gt_path = 'table-matching-ground-truth/ground-truth_revised/'
data_path = 'extracted-tables-revised/'
gdc_format_variable_names = set()

id_names = ['Case_ID', 'idx', 'Sample.ID',
            'case_id', 'SampleID', 'Participant']

possible_nulls = ['Not Reported',  'unknown', 'NaN', pd.NA, np.nan] 

for gt_file_name in os.listdir(gt_path):
    gt_file_path = os.path.join(gt_path, gt_file_name)
    if os.path.isfile(gt_file_path) and gt_file_name.endswith('.csv'):

        study = gt_file_name.replace('.csv', '')
        print(f"Study: {study}")

        dfs = []
        for data_file_name in os.listdir(data_path):
            # print(data_file_name)
            data_file_path = os.path.join(data_path, data_file_name)

            if os.path.isfile(data_file_path) and data_file_name.startswith(study) and data_file_name.endswith('.csv'):
                df = pd.read_csv(data_file_path)
                df['id'] = df.reset_index().index + 1
                dfs.append(df)
        # print(f"Ground Truth File: {gt_file_name} with {len(dfs)} data files")

        df_ground = pd.read_csv(gt_file_path)

        study_key = df_ground[df_ground['GDC_format_variable_names'] ==
                              'case_submitter_id']['original_paper_variable_names'].values[0]

        for target_col in df_ground['GDC_format_variable_names'].unique():

            if target_col not in intersection:
                continue

            study_col = df_ground[df_ground['GDC_format_variable_names']
                                  == target_col]['original_paper_variable_names'].values[0]
            # print(f"Study: {study_col}")

            dfs = [df for df in dfs if study_col in df.columns]

            if len(dfs) == 0:
                print(f"Target: {target_col} not found in any data files")
                continue
            if len(dfs) > 1:
                print(f"Target: {target_col} found in multiple data files")
                continue

            study_df = dfs[0]

            if study_key not in study_df.columns:
                continue

            # print(study_key)

            if study == 'Krug':
                # print(study_key)
                # print(study_df.head())
                if study_key in study_df.columns:
                    study_df['Sample.ID'] = study_df['Sample.ID'].str.replace(
                        '^X', '', regex=True)
                    


            partial_df_target = df_target[df_target["study"] == study]
            partial_df_target = partial_df_target[[
                'case_submitter_id', target_col, 'id']]

            partial_df_study = study_df[[study_key, study_col, 'id']]

            join_df = pd.merge(partial_df_study, partial_df_target,
                               left_on=study_key, right_on='case_submitter_id', how='inner')

            if join_df.shape[0] == 0:
                print(f"Joining {study} failed for column {target_col}")
                print(join_df.head())
                print('\n')


                # print(join_df.head())
                # print('\n')


            # print(join_df.head(5))
            # print('\n')
            file_name = f"{study}_{target_col}.csv"

            value_gt_path = 'value-matching-ground-truth/extracted/'
            join_df.to_csv(os.path.join(value_gt_path, file_name), index=False)

            value_gt_path = 'value-matching-ground-truth/deduplicated/'
            deuduplicaded_df = join_df.drop_duplicates()
            deuduplicaded_df.to_csv(os.path.join(value_gt_path, file_name), index=False)

            value_gt_path = 'value-matching-ground-truth/deduplicated_nonull/'    
            nonnull = join_df[~join_df.isin(possible_nulls).any(axis=1)]
            nonnull = nonnull.drop_duplicates()
            nonnull.to_csv(os.path.join(value_gt_path, file_name), index=False)


            # print(join_df.head(5))
            # print('\n')



```
