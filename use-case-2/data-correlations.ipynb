```python
import pandas as pd
import numpy as np
```

```python
def read_data(file_path, suffix=None):
    df = pd.read_csv(file_path, delimiter='\t')
    df = df.T
    if suffix:
        df = df.add_suffix(suffix)
    df = df.rename_axis('case_id').reset_index()
    return df
```

```python
file_path = './data/CPTAC/DISCOVERY_data_freeze_v_2.1/UCEC_discovery_Proteomics_PNNL_ratio_median_polishing_log2.txt'
df_proteo = read_data(file_path, suffix='_proteo')
df_proteo
```

```python
from itertools import product

sommut_count = df_sommut.sum(numeric_only=True)

proteins = list(df_proteo.columns)
proteins.remove('case_id')

smut = list(df_sommut.columns)
smut.remove('case_id')

count = 0
corrs = []

for protein, mutation in product(proteins, smut):
    corr = correlations[protein][mutation]
    mut_count = sommut_count[mutation]
    corrs.append((protein, mutation, corr, mut_count))
    if count == 100000000:
        break
    count += 1

df_corr = pd.DataFrame(columns=['protein', 'mutation', 'corr', 'sommut_count'], data=corrs)
```

```python
df_corr.sample(5)
```

#### Top correlations between genes/proteins
- *KRT26*: Diseases associated with KRT26 include *Anal Canal Squamous Cell Carcinoma* and Visceral Leishmaniasis. (https://www.genecards.org/cgi-bin/carddisp.pl?gene=KRT26)
- *HOXA4*: The involvement of HOXA4 in colorectal cancer and epithelial ovarian cancer has been reported. [...] In the current study, *HOXA4 expression was down-regulated* in lung cancer tissues when compared with non-cancerous tissues. (https://www.nature.com/articles/s41419-018-0497-x)
- *GAGE12J*: mediates human gastric carcinoma growth and metastasis (https://pubmed.ncbi.nlm.nih.gov/25346337/)

```python
df_corr['rank'] = df_corr['corr'].abs() * (1.0 - 1.0/np.sqrt(df_corr['sommut_count']-3))
df_corr_sorted  = df_corr.sort_values(by='rank', ascending=False)
df_corr_sorted.head(60)
```
