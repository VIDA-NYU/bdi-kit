# Integration with Claude Desktop

1. To integrate bdi-kit with AI assistants (e.g., Claude) via via Model Context Protocol (MCP), install it using [uv](https://docs.astral.sh/uv/):

    ```bash
        uv tool install bdi-kit[mcp]

2. Add an entry to the Claude Desktop settings (Settings -> Developer -> Edit Config) $^{*}$. For details on locating and editing the Claude Desktop settings, see [this example](https://modelcontextprotocol.io/quickstart/user#installing-the-filesystem-server) in the official Claude documentation.

    ```json
    {
      "mcpServers": {
        "bdikit-harmonizer": {
          "command": "bdikit-mcp"
        }
      }
    }
    ```

Depending on where Claude resolves executables, you may need to specify the full path.
Instead of `"command": "bdikit-mcp"` use `"command": "YOUR_PATH/bdikit-mcp"`. You can find `YOUR_PATH` by running `which bdikit-mcp`.

3. After restarting Claude Desktop, bdi-kit should appear as an available tool by clicking on <img style="display: inline; margin: 0px; height: 1.3em;" src="https://mintlify.s3.us-west-1.amazonaws.com/mcp/images/claude-desktop-mcp-slider.svg"> (see the image below).

<img src="../docs/source/_static/images/available_tools.png" alt="BDI-Kit as available tool" style="width: 500px;"/>

After these steps, you can start the harmonization process directly in the chat interface. For example, you can type: "Harmonize the schema of a dataset located at '/home/my_dataset.csv' using bdi-kit"

$^{*}$ For primitives that depend on LLMs, you must specify the model using the `llm` argument. In some cases, you will also need to provide your API key through the `env` field, e.g. for OpenAI models:


```json
    {
      "mcpServers": {
        "bdikit-harmonizer": {
          "command": "bdikit-mcp",
          "args": ["--llm", "YOUR_LLM"],
          "env": {
              "OPENAI_API_KEY": "YOUR_KEY"
            }
        }
      }
    }
```
