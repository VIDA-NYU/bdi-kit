# Integration with Claude Desktop

1. To use bdi-kit from AI agents or assistants (e.g. Claude) via Model Context Protocol (MCP), install:

    ```bash
    pip install bdi-kit[mcp]

2. Download the [mcp_server.py](https://github.com/VIDA-NYU/bdi-kit/blob/devel/examples/mcp_server.py) module. It contains just two lines of code that start the MCP server:


    ```python
    from bdikit.mcp import server
    server.run(transport="stdio")

3. Add an entry to the Claude Desktop settings (JSON file) named "bdikit-harmonizer". The entry runs `mcp_server.py` by Python $^{*}$, make sure the path to this file is correct. For details on locating and editing the Claude Desktop settings, see [this example](https://modelcontextprotocol.io/quickstart/user#installing-the-filesystem-server) in the official Claude documentation.

    ```json
    {
      "mcpServers": {
        "bdikit-harmonizer": {
          "command": "bash",
          "args": [
            "-c",
            "python YOUR_PATH/mcp_server.py"
          ]
        }
      }
    }
    ```

4. After restarting Claude Desktop, bdi-kit should appear as an available tool by clicking on <img style="display: inline; margin: 0px; height: 1.3em;" src="https://mintlify.s3.us-west-1.amazonaws.com/mcp/images/claude-desktop-mcp-slider.svg"> (see the image below).

<img src="../docs/source/_static/images/available_tools.png" alt="BDI-Kit as available tool" style="width: 500px;"/>

5. Start the harmonization process directly in the chat interface. For example, you can type: "Harmonize the schema of a dataset located at '/home/my_dataset.csv' using bdi-kit"

$^{*}$ For certain primitives, such as `llm` and `llm_numeric`, you must provide your OpenAI API key. If it is already set in your environment, you can ignore this note. Otherwise, you can pass it as an environment variable when starting the MCP server, for example: `export OPENAI_API_KEY='YOUR_KEY' && python YOUR_PATH_/mcp_server.py`
