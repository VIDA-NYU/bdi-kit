# BDIViz Result Analysis

```python
import bdikit as bdi
import pandas as pd
```

```python
#For yurong: Add SANTOS/STARMIE/PLAIN ENCODING here

def benchmark_accuracy(df: pd.DataFrame, ground_truth: pd.DataFrame, method: str = "coma"):
    columns = ground_truth["original_paper_variable_names"].values
    if method == "bdiviz":
        column_mappings = df
    else:
        column_mappings = bdi.match_schema(df[columns], target="gdc", method=method)

    count = 0
    correct_count = 0
    for index, row in ground_truth.iterrows():
        gt_source = row["original_paper_variable_names"]
        gt_target = row["GDC_format_variable_names"]
        if column_mappings[column_mappings["source"] == gt_source]["target"].values[0] in [x.strip() for x in gt_target.split(";")]:
            correct_count += 1
        count += 1

    return correct_count / count # Reture a accuracy rate for every dataset
```

```python
from pathlib import Path

df_dict = {
    "Matcher": [],
    "Dataset": [],
    "Accuracy": [],
}
pathlist = list(Path("./groundtruth").glob('*.csv'))

for path in pathlist:
    # because path is object not string
    path_in_str = str(path)   
    file_name = path_in_str.split("/")[1]
    df_path = f"./datasets/{file_name}"
    
    df = pd.read_csv(df_path)
    ground_truth = pd.read_csv(path_in_str)
    
    for method in ["two_phase"]: #["coma", "similarity_flooding", "cupid", "distribution_based", "jaccard_distance"]:
        print(f"{method}-{file_name}")
        accuracy = benchmark_accuracy(df, ground_truth, method=method)
        df_dict["Matcher"].append(method)
        df_dict["Dataset"].append(file_name.split(".")[0])
        df_dict["Accuracy"].append(accuracy)


df = pd.DataFrame(df_dict)
df
```

```python
import altair as alt

chart = alt.Chart(df).mark_bar().encode(
        x='Matcher',
        y='Accuracy',
        column='Dataset',
        color='Matcher',
    )

            
chart         
```
