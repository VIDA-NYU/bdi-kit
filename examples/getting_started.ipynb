```python
column_mappings = pd.DataFrame(
    [
        {
            "source": "Ethnicity",
            "target": "ethnicity",
        },
        {
            "source": "Gender",
            "target": "gender",
        },
        {
            "source": "Race",
            "target": "race",
        },
        {
            "source": "Country",
            "target": "country_of_birth",
        },
        {
            "source": "Tumor_Focality",
            "target": "tumor_focality",
        },
        {
            "source": "FIGO_stage",
            "target": "figo_stage",
        },
        {
            "source": "Histologic_Grade_FIGO",
            "target": "tumor_grade",
        },
        {
            "source": "BMI",
            "target": "bmi",
        }
    ]
)
```

```python
mappings = bdi.preview_value_mappings(
    dataset,
    column_mapping=column_mappings,
    target="gdc",
    method="edit_distance",
)

for mapping in mappings:
    print(f"{mapping['source']} => {mapping['target']}")
    display(mapping["mapping"])
    print("")
```

```python
value_mappings = bdi.match_values(
    dataset, target="gdc", column_mapping=column_mappings, method="edit_distance"
)
```

```python
from bdikit import update_mappings
import re


def map_hist_grade_figo(value_str):
    if isinstance(value_str, str):
        return re.sub(r"FIGO grade (\d+)", r"G\1", value_str)
    else:
        return "Unknown"


user_mappings = [
    {
        "source": "BMI",
        "target": "bmi",
    },
    {
        "source": "Age",
        "target": "days_to_birth",
        "mapper": lambda age: -age * 365,
    },
    {
        "source": "Histologic_Grade_FIGO",
        "target": "tumor_grade",
        "mapper": map_hist_grade_figo,
    },
    {
        "source": "FIGO_stage",
        "target": "figo_stage",
        "mapper": lambda x: f"Stage {x}" if isinstance(x, str) else "Unknown",
    },
]

harmonization_spec = update_mappings(value_mappings, user_mappings)
```

```python
harmonized_dataset = bdi.materialize_mapping(dataset, harmonization_spec)
harmonized_dataset
```

```python
original_columns = map(lambda m: m["source"], harmonization_spec)
dataset[original_columns]
```
