## Viewing Value Match Results

```python
import bdikit as bdi
import pandas as pd
```

In this example, we are mapping data from Dou et al. (https://pubmed.ncbi.nlm.nih.gov/37567170/) to the GDC format.

```python
dataset = pd.read_csv('./datasets/dou.csv')
columns = [
    "Country",
    "Histologic_type",
    "FIGO_stage",
    "Race",
    "Ethnicity",
]

dataset[columns].head(10)
```

View results of value matching for a single column:

```python
column_mapping = ("FIGO_stage", "figo_stage")

value_mappings = bdi.match_values(
        dataset,
        target="gdc",
        column_mapping=column_mapping,
        method="tfidf",
    )
```

```python
bdi.view_value_matches(value_mappings)
```

View results of value matching for multiple columns:

```python
column_mappings = bdi.match_schema(dataset[columns], target="gdc", method="two_phase")
column_mappings
```

```python
value_mappings = bdi.match_values(
        dataset,
        target="gdc",
        column_mapping=column_mappings,
        method="tfidf",
    )
```

```python
bdi.view_value_matches(value_mappings)
```
