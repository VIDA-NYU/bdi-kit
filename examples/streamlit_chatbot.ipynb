# Using bdi-kit Through a Chatbot Agent

1. To use bdi-kit via a simple chatbot-like application, install it using [uv](https://docs.astral.sh/uv/):

    ```bash
        uv tool install bdi-kit[chatbot]

2. Run the following command:

    ```bash
        bdikit-chatbot --llm YOUR_LLM

Where `YOUR_LLM` is the model to be used by the agent and some LLM-based primitives in bdi-kit (e.g. `ollama/deepseek-r1:32b`). In some cases, you will also need to provide your API key through environment variables.

After these steps, you can start the harmonization process directly in the chat interface. For example, you can type: "Harmonize the schema of a dataset located at '/home/my_dataset.csv' using bdi-kit" (see the image below).
