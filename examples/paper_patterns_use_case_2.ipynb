# Use Case 2: Table-to-Model Harmonization

First, import the BDI-Kit library and other libraries.

```python
import bdikit as bdi
import pandas as pd
```

## Loading the data

The dataset used in this example cannot be released due to privacy restrictions. It can be accessed through Synapse at https://www.synapse.org/Synapse:syn62827833, but requires submitting an access request via the Sage Bionetworks support portal at https://sagebionetworks.jira.com/servicedesk/customer/portals 

```python
source_dataset = pd.read_csv("./datasets/GEO_Metadata.csv")
source_dataset.head(5)
```

## Schema Matching

```python
column_mappings = bdi.rank_schema_matches(source_dataset, columns=source_dataset.columns, target="synapse", top_k=4, method="max_val_sim", method_args={"contribution_factor": 1}, standard_args={"subschema_name": "RNASeqTemplate"})
column_mappings
```

## Evaluation of Matches

```python
explanations = bdi.evaluate_schema_matches(column_mappings, source_dataset, target="synapse", standard_args={"subschema_name": "RNASeqTemplate"})
explanations
```
